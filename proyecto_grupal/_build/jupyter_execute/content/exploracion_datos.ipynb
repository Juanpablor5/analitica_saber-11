# Exploración de Datos

La exploración de datos es el proceso inicial de análisis en el que se utilizan técnicas estadísticas y visuales para comprender mejor la estructura y características de un conjunto de datos. Esto incluye la identificación de tipos de variables, la evaluación de la calidad de los datos mediante la detección de valores nulos o atípicos, y el cálculo de estadísticas descriptivas como medias y desviaciones estándar. Además, se emplean gráficos como histogramas y diagramas de dispersión para identificar patrones, tendencias y relaciones entre variables. Este proceso es fundamental para obtener una visión general de los datos, facilitando la formulación de hipótesis y la planificación de análisis más profundos.

*Importación de Librerías*

Las librerías utilizadas para la exploración de los datos en python son:

`pandas`
`matplotlib`
`seaborn`


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Se realiza la importación del archivo con los datos completos descargado de https://www.datos.gov.co/Educaci-n/Resultados-nicos-Saber-11/kgxf-xxbe/about_data 

```python
file_path = "Resultados__nicos_Saber_11_20240528.csv"
df = pd.read_csv(file_path)
```

Los datos se leen como un DataFrame, revelando un total de 51 columnas que incluyen tanto datos numéricos como categóricos. Este conjunto de datos proporciona una variedad de información detallada, abarcando diferentes aspectos de los resultados académicos y las características sociodemográficas de los estudiantes.

## Resumen de los Datos Numéricos:

- *PUNT_INGLES:* Los puntajes en inglés varían entre 0 y 100 con una media de 53.27.

- *PUNT_MATEMATICAS:* Los puntajes en matemáticas también varían entre 0 y 100 con una media de 52.9.

- *PUNT_SOCIALES_CIUDADANAS:* Los datos de puntajes en sociales y ciudadanas están incompletos, con un conteo de solo 103, su media es de 52.9.

- *PUNT_C_NATURALES y PUNT_LECTURA_CRITICA:* También tienen datos incompletos, con 103 registros, su media es de 53.87 y 55.64 respectivamente.

- *PUNT_GLOBAL:* Presenta un promedio de 296.51 entre los 103 registros disponibles.

## Resumen de los Datos Categóricos:
- *COLE_CALENDARIO:* La mayoría de los colegios siguen el calendario A.

- *COLE_GENERO:* La mayoría de los colegios son mixtos.

- *COLE_JORNADA:* Predomina la jornada Completa.

- *FAMI_EDUCACIONMADRE y FAMI_EDUCACIONPADRE:* La mayoría de los padres tienen educación secundaria completa.

- *FAMI_ESTRATOVIVIENDA:* Predomina el estrato 2.

- *FAMI_TIENEAUTOMOVIL, FAMI_TIENECOMPUTADOR y FAMI_TIENEINTERNET:* Una mayoría significativa de estudiantes no tienen automóvil, pero sí tienen computadora e internet en sus hogares

## Identificación de Datos Vacíos

Se identificaron datos vacíos especialmente en las variables de puntaje global, puntaje ciencias naturales, puntaje de sociales y  puntaje lectura crítica.

También se identificó un gran numero de datos vacíos en la variable categorica colegio bilingüe.

```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.figure(figsize=(12, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Valores nulos en el DataFrame')
plt.show()
```

## Distribución de puntaje Global 

Podemos ver que la variable puntaje global parece tener un comportamiento normal centrada en la media de 269.51

```python
# Análisis de distribución de una variable numérica: Puntajes en Inglés
plt.figure(figsize=(10, 6))
sns.histplot(df['PUNT_GLOBAL'], kde=True, bins=30)
plt.title('Distribución puntaje Global')
plt.xlabel('Puntaje Global')
plt.ylabel('Frecuencia')
plt.show()
```

## Proporciones de Variables categóricas

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='COLE_JORNADA')
plt.title('Frecuencia de las Jornadas del Colegio')
plt.xlabel('Jornada')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
df_melted = df.melt(value_vars=['FAMI_EDUCACIONMADRE', 'FAMI_EDUCACIONPADRE'], 
                    var_name='Parentezco', value_name='Nivel_Educacion')

sns.countplot(data=df_melted, x='Nivel_Educacion', hue='Parentezco', palette='viridis')
plt.title('Frecuencia de los Niveles Educativos de Madre y Padre')
plt.xlabel('Nivel Educativo')
plt.ylabel('Frecuencia')
plt.xticks(rotation=90)
plt.legend(title='Parentezco')
plt.show()
```
